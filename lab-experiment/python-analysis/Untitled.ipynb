```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(content.Mois_beans)
axes[0, 0].set_title('Moisture Content of Beans')
axes[0, 0].tick_params(axis='x', rotation=90)

axes[0, 1].plot(content.Mois_Maize, 'r')
axes[0, 1].set_title('Moisture Content of Maize')
axes[0, 1].tick_params(axis='x', rotation=90)

axes[1, 0].plot(content.Mois_Onion, 'g')
axes[1, 0].set_title('Moisture Content of Onion')
axes[1, 0].tick_params(axis='x', rotation=90)

axes[1, 1].plot(content.Mois_Rice, 'y')
axes[1, 1].set_title('Moisture Content of Rice')
axes[1, 1].tick_params(axis='x', rotation=90)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Moisture_Content.png')
# Show the plot
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(content.Temp_beans)
axes[0, 0].set_title('Temperature of Beans')
axes[0, 0].tick_params(axis='x', rotation=90)

axes[0, 1].plot(content.Temp_Maize, 'r')
axes[0, 1].set_title('Temperature of Maize')
axes[0, 1].tick_params(axis='x', rotation=90)

axes[1, 0].plot(content.Temp_Onion, 'g')
axes[1, 0].set_title('Temperature of Onion')
axes[1, 0].tick_params(axis='x', rotation=90)

axes[1, 1].plot(content.Temp_Rice, 'y')
axes[1, 1].set_title('Temperature of Rice')
axes[1, 1].tick_params(axis='x', rotation=90)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Temperature.png')
# Show the plot
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(content.Humid_beans)
axes[0, 0].set_title('Humidity of Beans')
axes[0, 0].tick_params(axis='x', rotation=90)

axes[0, 1].plot(content.Humid_Maize, 'r')
axes[0, 1].set_title('Humidity of Maize')
axes[0, 1].tick_params(axis='x', rotation=90)

axes[1, 0].plot(content.Humid_Onion, 'g')
axes[1, 0].set_title('Humidity of Onion')
axes[1, 0].tick_params(axis='x', rotation=90)

axes[1, 1].plot(content.Humid_Surr, 'y')
axes[1, 1].set_title('Humidity of the Surrounding')
axes[1, 1].tick_params(axis='x', rotation=90)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Humidity.png')
# Show the plot
plt.show()
```
