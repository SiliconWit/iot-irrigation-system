```python
import matplotlib.pyplot as plt
import numpy as np

# Sample water usage data for each crop and growth stage
beans_data = [452.33, 103.9, 0, 0]
maize_data = [432.15, 76, 0, 0]
onion_data = [17.55, 36.04, 0, 0]
rice_data = [ 403.57,  0, 0, 0]

# Growth stages
growth_stages = ['Initial', 'Development', 'Mid-Season', 'Late-Season']

# Set up plot positions
bar_width = 0.2
index = np.arange(len(growth_stages))

# Plotting the bars
plt.plot(index, beans_data, marker='o', label='Estimated Water')
plt.plot(index, maize_data, marker='o', label='Utilized Water')
# plt.plot(index, onion_data, marker='o', label='Utilized Water (onion)')
# plt.plot(index, rice_data, marker='o', label='Utilized Water (rice)')

for i, value in enumerate(beans_data):
    plt.text(i, value + 0.5, str(value), ha='center')
for i, value in enumerate(maize_data):
    plt.text(i, value + 0.5, str(value), ha='center')
# for i, value in enumerate(onion_data):
#     plt.text(i, value + 0.5, str(value), ha='center')
# for i, value in enumerate(rice_data):
#     plt.text(i, value + 0.5, str(value), ha='center')


plt.xlabel('Growth Stages')
plt.ylabel('Water Usage (mm)')
plt.title('Water Comparison (37-Experimental Days)')
plt.xticks(index, growth_stages)
plt.legend()
plt.grid()
plt.savefig('Actual Water')
# Show the plot
plt.show()
```
