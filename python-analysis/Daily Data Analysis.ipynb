```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

%matplotlib notebook
```

```python
Data = pd.read_csv('Daily.csv', index_col=0, parse_dates= True)
```

```python
Data
```

```python
Data.describe()
```

```python
%matplotlib inline
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Data.Mois_beans)
plt.title('Moisture content in Beans block')
plt.xlabel('Period (Days)')
plt.xticks(rotation=0)
plt.ylabel('Moisture content (%)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
#plt.hist(content.Day, content.water_level, bins=10,color='skyblue', edgecolor='black')
plt.plot(Data.water_level,color='g')
plt.xlabel('Period (Days)')
plt.xticks(rotation=0)
plt.ylabel('Level (mm)')
plt.title('Water Level of Rice block')
plt.savefig('Daily Water Level Rice Block')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(Data.Temp_beans)
plt.xlabel('Period (Days)')
plt.xticks(rotation=0)
plt.ylabel('Temp.Beans')
plt.title('Temperature of Beans block')
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(Data.Mois_beans)
axes[0, 0].set_title('Moisture Content of Beans')
axes[0, 0].set_xlabel('Period (Days)')
axes[0, 0].set_ylabel('Moisture (%)')
axes[0, 0].tick_params(axis='x', rotation=0)

axes[0, 1].plot(Data.Mois_Maize, 'r')
axes[0, 1].set_title('Moisture Content of Maize')
axes[0, 1].set_xlabel('Period (Days)')
axes[0, 1].set_ylabel('Moisture (%)')
axes[0, 1].tick_params(axis='x', rotation=0)

axes[1, 0].plot(Data.Mois_Onion, 'g')
axes[1, 0].set_title('Moisture Content of Onion')
axes[1, 0].set_xlabel('Period (Days)')
axes[1, 0].set_ylabel('Moisture (%)')
axes[1, 0].tick_params(axis='x', rotation=0)

axes[1, 1].plot(Data.Mois_Rice, 'y')
axes[1, 1].set_title('Moisture Content of Rice')
axes[1, 1].set_xlabel('Period (Days)')
axes[1, 1].set_ylabel('Moisture (%)')
axes[1, 1].tick_params(axis='x', rotation=0)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Daily Moisture_Content.png')
# Show the plot
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(Data.Temp_beans)
axes[0, 0].set_title('Temperature of Beans')
axes[0, 0].set_xlabel('Period (Days)')
axes[0, 0].set_ylabel('Temp. (oC)')
axes[0, 0].tick_params(axis='x', rotation=0)

axes[0, 1].plot(Data.Temp_Maize, 'r')
axes[0, 1].set_title('Temperature of Maize')
axes[0, 1].set_xlabel('Period (Days)')
axes[0, 1].set_ylabel('Temp. (oC)')
axes[0, 1].tick_params(axis='x', rotation=0)

axes[1, 0].plot(Data.Temp_Onion, 'g')
axes[1, 0].set_title('Temperature of Onion')
axes[1, 0].set_xlabel('Period (Days)')
axes[1, 0].set_ylabel('Temp. (oC)')
axes[1, 0].tick_params(axis='x', rotation=0)

axes[1, 1].plot(Data.Temp_Rice, 'y')
axes[1, 1].set_title('Temperature of Rice')
axes[1, 1].set_xlabel('Period (Days)')
axes[1, 1].set_ylabel('Temp. (oC)')
axes[1, 1].tick_params(axis='x', rotation=0)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Daily Temperature.png')
# Show the plot
plt.show()
```

```python
# Create a figure and a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(10, 5))

# Plot data in each subplot and rotate x-axis labels
axes[0, 0].plot(Data.Humid_beans)
axes[0, 0].set_title('Humidity of Beans')
axes[0, 0].set_xlabel('Period (Days)')
axes[0, 0].set_ylabel('Humid. (%)')
axes[0, 0].tick_params(axis='x', rotation=0)

axes[0, 1].plot(Data.Humid_Maize, 'r')
axes[0, 1].set_title('Humidity of Maize')
axes[0, 1].set_xlabel('Period (Days)')
axes[0, 1].set_ylabel('Humid. (%)')
axes[0, 1].tick_params(axis='x', rotation=0)

axes[1, 0].plot(Data.Humid_Onion, 'g')
axes[1, 0].set_title('Humidity of Onion')
axes[1, 0].set_xlabel('Period (Days)')
axes[1, 0].set_ylabel('Humid. (%)')
axes[1, 0].tick_params(axis='x', rotation=0)

axes[1, 1].plot(Data.Humid_Surr, 'y')
axes[1, 1].set_title('Humidity of the Surrounding')
axes[1, 1].set_xlabel('Period (Days)')
axes[1, 1].set_ylabel('Humid. (%)')
axes[1, 1].tick_params(axis='x', rotation=0)
# Adjust spacing between subplots
plt.tight_layout()
plt.savefig('Daily Humidity.png')
# Show the plot
plt.show()
```
